# Required libraries

```python
from random import randint
from copy import deepcopy
from IPython.display import clear_output
from numpy import array
```

# Generator

```python
def createLetter(matrix, letter, coordinates, symbol):
    for x, y in coordinates: 
        # upper part
        matrix[x][y] = symbol
        matrix[x][y + 1] = symbol
        matrix[x][y + 2] = symbol

        # upper part (second row)
        matrix[x + 1][y] = symbol
        matrix[x + 1][y + 2] = symbol

        # upper part (third row)
        matrix[x + 2][y] = symbol
        matrix[x + 2][y + 1] = symbol
        matrix[x + 2][y + 2] = symbol     
        
    
        if letter == 'P':
            # other parts of letter
            matrix[x+ 3][y] = symbol
            matrix[x+ 4][y] = symbol
        
        elif letter == 'q':
            # other parts of letter
            matrix[x+ 3][y + 2] = symbol
            matrix[x+ 4][y + 2] = symbol
    
    return matrix



# to check if new coordinates can be placed
def check(indexes, currentIndex, letter):
    
    if(letter == 'P'):

        # to check every P placed before
        for x, y in indexes:

            # checking with left side of P
            if currentIndex[1] == y: 
                if currentIndex[0] in [i for i in range(x, x + 5)]:
                    return False

            # checking middle and right part of P
            elif currentIndex[1] in [i for i in range(y + 1, y + 3)]:
                if currentIndex[0] in [i for i in range(x, x + 3)]:
                    return False

            
            # also checking if any of indexes is in area of current index
            if y == currentIndex[1]:
                if x in [i for i in range(currentIndex[0], currentIndex[0] + 5)]:
                    return False

            elif y in [i for i in range(currentIndex[1] + 1, currentIndex[1] + 3)]:
                if x in [i for i in range(currentIndex[0], currentIndex[0] + 3)]:
                    return False



    elif(letter == 'q'):

        # to check every q placed before
        for x, y in indexes:

            # checking with left side of q
            if currentIndex[1] in [i for i in range(y, y + 3)]:
                if currentIndex[0] in [i for i in range(x, x + 3)]:
                    return False

            # checking middle and right part of q
            elif currentIndex[1] == y + 2: 
                if currentIndex[0] in [i for i in range(x, x + 5)]:
                    return False

            
            # also checking if any of indexes is in area of current index
            if y in [i for i in range(currentIndex[1], currentIndex[1] + 3)]:
                if currentIndex[0] in [i for i in range(currentIndex[0], currentIndex[0] + 3)]:
                    return False
            
            elif y == currentIndex[1] + 2:
                if x in [i for i in range(currentIndex[0], currentIndex[0] + 5)]:
                    return False

    return True


def coordinateGenerator(num, width, height, letter):

    result= []
        
    # random generating
    while(len(result) != num):
        x, y= randint(0, height - 5 - 1), randint(0, width - 5 - 1) # randint inclueds both start and end
        # print(x, y)
        if(check(result, (x, y), letter)):
            result.append((x, y))

    return result


def noiseMatrix(matrix, symbol):
    num= (len(matrix) * len(matrix[0])) // 5  # noising value

    for _ in range(num):
        x, y= randint(0, len(matrix) - 1), randint(0, len(matrix[0]) - 1)
        matrix[x][y] = symbol

    return matrix


def generate(num, letter, width, height):

    symb= '.'
    symb2= 'O'
    matrix= []

    # creating area with same symbol
    for _ in range(height):
        matrix.append([symb] * width)

    # creating letters in matrix
    coordinates= coordinateGenerator(num, width, height, letter)
    matrix = createLetter(matrix, letter, coordinates, symb2)

    # noising matrix to hide letters
    matrix2= deepcopy(matrix)
    matrix= noiseMatrix(matrix, symb2)

    return matrix, matrix2
```

# Detector

In this method, firstly we create new matrix which is our letter(5x3 size), and then we will test it with submatrices of orginal matrix if they both are same or not.

```python
def printMatrix(matrix):
    for x in matrix:
        print(x)

def equalityCheck(matrix, letterMatrix, letter):
    
    # we should ignore some bottom parts

    if letter == 'P':
        for x in range(len(matrix)):
            for y in range(len(matrix[0])):
                if x in [3, 4] and y in [1, 2]:
                    continue
                if matrix[x][y] != letterMatrix[x][y]:
                    return False

    if letter == 'q' :
        for x in range(len(matrix)):
            for y in range(len(matrix[0])):
                if x in [3, 4] and y in [0, 1]:
                    continue
                if matrix[x][y] != letterMatrix[x][y]:
                    return False

    return True

def find(matrix, letter):

    # Printing matrix itself
    print('Orginal Matrix:')
    printMatrix(matrix)


    # creating letter matrix
    symb= '.'
    symb2= 'O'
    letterMatrix= []

    for _ in range(5): 
        letterMatrix.append([symb] * 3)

    letterMatrix = createLetter(letterMatrix, letter, [(0, 0)], symb2)

    # checking part
    letterCoordinates= []
    for x in range(len(matrix) - 4):
        for y in range(len(matrix[0]) - 2):
            if equalityCheck(list(array(matrix)[x:x+5, y:y+3]), letterMatrix, letter):
                letterCoordinates.append((x, y))

    # printing results
    print(f'\nThe count of letter: {len(letterCoordinates)}')
    print('The coordinates: ')
    printMatrix(letterCoordinates)
```

```python
# defining variables
height, width= 10, 10
letter= 'P'
num= 2

flag= 'n'
while flag == 'n':
    matrix, matrix2= generate(num, letter, width, height)

    print('Matrix before noising: ')
    printMatrix(matrix2)

    flag= input('Are you satisfied with matrix? ')
    clear_output(wait=True)

print('Matrix before noising: ')
printMatrix(matrix2)
print()

if flag == 'y':
    find(matrix, letter)
```

# GAN model :D

This simple Generative Adversarial Network model is created to get rid of bug of generator.

For this purpose, we have to change 'Find' function to get results as returned value (we don't want function to print result)

```python
def findGAN(matrix, letter):
    
    # creating letter matrix
    symb= '.'
    symb2= 'O'
    letterMatrix= []

    for _ in range(5): 
        letterMatrix.append([symb] * 3)

    letterMatrix = createLetter(letterMatrix, letter, [(0, 0)], symb2)

    # checking part
    letterCoordinates= []
    for x in range(len(matrix) - 4):
        for y in range(len(matrix[0]) - 2):
            if equalityCheck(list(array(matrix)[x:x+5, y:y+3]), letterMatrix, letter):
                letterCoordinates.append((x, y))

    # returning coordinates
    return letterCoordinates
```

```python
def processGAN(height, width, letter, num):
    print('Generating...')
    matrix, matrix2= generate(num, letter, width, height)
    count = len(findGAN(matrix2, letter))
    return matrix, matrix2, count


def ganModel(height, width, letter, num):
    count = -1
    while count != num:
        matrix, matrix2, count= processGAN(height, width, letter, num)
    return matrix, matrix2
    

height, width, letter, num= 10, 15, 'q', 4
matrix, matrix2 = ganModel(height, width, letter, num)

printMatrix(matrix2)
```

```python
def runModel(height, width, letter, num):
    matrix, matrix2 = ganModel(height, width, letter, num)
    print('Matrix before noising: ')
    printMatrix(matrix2)
    find(matrix, letter)
```

```python
height, width, letter, num= 15, 15, 'P', 4
runModel(height, width, letter, num)
```
